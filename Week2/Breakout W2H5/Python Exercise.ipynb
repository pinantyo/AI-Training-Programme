1. Membuat fungsi mencetak "Belajar Python itu menyenangkan" 10x

```python
def cetak_10x():
  for i in range(10):
    print("Belajar Python itu menyenangkan")

cetak_10x()
```

2. nama fungsi: cetak_apapun_10x 
terdapat satu argument: teks 
tidak mengembalikan nilai 
tidak ada input dari pengguna 

```python
def cetak_apapun_10x(teks):
  for i in range(10):
    print(teks)

cetak_apapun_10x("Belajar pemrograman dapat mengasah logika berfikir")
```

3. nama fungsi: 
cetak_apapun_10x_dengan_input 
tidak ada argument 
tidak mengembalikan nilai 
terdapat input dari pengguna 

```python
def cetak_apapun_10x_dengan_input():
  teks = input("")
  for i in range(10):
    print(teks)
    

cetak_apapun_10x_dengan_input()
```

4. nama fungsi: cetak_apapun_10x_dengan_default 
terdapat argument default: teks 
value dari argument default: "Selamat belajar python, jangan lupa perbanyak latihan yaa" tidak mengembalikan nilai 
tidak ada input dari pengguna 


```python
def cetak_apapun_10x_dengan_default(teks = "Selamat belajar python, jangan lupa perbanyak latihan yaa"):
  for i in range(10):
    print(teks)

cetak_apapun_10x_dengan_default()
cetak_apapun_10x_dengan_default("Oh senangnya, aku senang sekali :D")
```
