> Buatlah game suit batu-gunting-kertas sederhana. Di dalam program tersebut terdapat dua orang pemain yang dapat memilih batu/gunting/kertas menggunakan input dari pengguna. Untuk mempermudah penggambaran aturan pemenang dalam game, lengkapilah tabel kemenangan dengan menggunakan semua kemungkinan yang tersedia.

```python
game = ["kertas","batu","gunting"]
Player1 = input("Masukan pilihan pemain 1: ")
Player2 = input("Masukan pilihan pemain 2: ")

if(Player1.lower() in game and Player2.lower() in game):
  if(Player1 == Player2):
    print("Seri")
  elif((Player1 == "kertas" and Player2 == "batu") or (Player1 == "gunting" and Player2 == "kertas") or (Player1 == "batu" and Player2 == "gunting")):
    print("Pemain 1")
  else:
    print("Pemain 2")
else:
  print("Input tidak valid!. Gunakan \"batu\", \"gunting\" dan \"kertas\" tanpa tanda kutip!")
```

```python
#Test Case
Pemain1 = Pemain2 = ["kertas","batu","gunting"]
for Player1 in Pemain1:
  for Player2 in Pemain2:
    if(Player1.lower() in game and Player2.lower() in game):
      if(Player1 == Player2):
        print("Seri")
      elif((Player1 == "kertas" and Player2 == "batu") or (Player1 == "gunting" and Player2 == "kertas") or (Player1 == "batu" and Player2 == "gunting")):
        print("Pemain 1")
      else:
        print("Pemain 2")
    else:
      print("Input tidak valid!. Gunakan \"batu\", \"gunting\" dan \"kertas\" tanpa tanda kutip!")
```
